```python
%matplotlib inline
%pylab inline
import skimage.io

pylab.rcParams['figure.figsize'] = (10, 10)
```

```python
img = skimage.io.imread('duckling.png').astype('float64') / 256.0
skimage.io.imshow(img)
```

```python
print img.shape
```

```python
width = img.shape[0]
length = img.shape[1]

left_half = img[:, :length / 2, :]
skimage.io.imshow(left_half)
skimage.io.imsave('left_half.png', left_half)
```

```python
top_half = img[:width / 2, :, :]
skimage.io.imshow(top_half)
skimage.io.imsave('top_half.png', top_half)
```

```python
import copy
from matplotlib import pyplot as plt

channels = []

for a in xrange(3):
    plt.figure()
    temp = img.copy()
    temp[:, :, a] = 0
    skimage.io.imshow(temp)
    channels.append(temp)
```

    .

```python
channels = []

for a in xrange(3):
    plt.figure()
    temp = copy.deepcopy(img)
    t = range(3)
    t.remove(a)
    t = tuple(t)
    temp[:, :, t] = 0
    skimage.io.imshow(temp)
    channels.append(temp)
```

```python
import numpy as np
import skimage.color

gray_img = skimage.color.rgb2gray(img)
gray_mean_img = np.mean(img[: ,: , :3], axis=2)

skimage.io.imshow(np.concatenate((gray_mean_img, gray_img), axis=1))
```

   ,   ,     .

```python
img_hsv = skimage.color.rgb2hsv(img[:, :, :3])
skimage.io.imshow(img_hsv)
```

```python
for a in xrange(3):
    plt.figure()
    temp = img_hsv.copy()
    t = range(3)
    t.remove(a)
    t = tuple(t)
    temp[:, :, t] = 0
    skimage.io.imshow(temp)
```

```python
inverted_img = 1.0 - img
skimage.io.imshow(skimage.color.rgb2gray(inverted_img))
```

```python
inverted_img = img[:, ::-1, :]
skimage.io.imshow(inverted_img)
```
